# Comparison of local partitioning of atomic contacts

This jupyter notebook allows to reproduce the results published in the following paper:
- Lorenza Pacini, Claire Lesieur, A computational methodology to diagnose sequence-variant dynamic perturbations by comparing atomic protein structures, Bioinformatics, Volume 38, Issue 3, 1 February 2022, Pages 703â€“709, https://doi.org/10.1093/bioinformatics/btab736

```python
import os
import tools
from Protein import Protein
```

```python
results_folder = "results"
os.makedirs(results_folder, exist_ok=True)
data_folder = "data"
```

## AB5 toxins

```python
name1 = 'CtxB5'
pdb_id_1 = "1eei" # CtxB5
pdb_path_1 = os.path.join(data_folder, pdb_id_1 + ".pdb")
name2 = 'HeatLabile'
pdb_id_2 = "1ltr" # LTB5
pdb_path_2 = os.path.join(data_folder, pdb_id_2 + ".pdb")
results_subfolder = os.path.join(results_folder, "AB5toxins")
os.makedirs(results_subfolder, exist_ok=True)

pos_start = 1
pos_stop = 102
```

```python
P1 = Protein(pdb_path_1, pos_start=pos_start, pos_stop=pos_stop)
P2 = Protein(pdb_path_2, pos_start=pos_start, pos_stop=pos_stop)
```

```python
tools.plot_local_partitioning_difference(P1, P2, results_subfolder, name="%s_vs_%s" %(name2, name1))
```
